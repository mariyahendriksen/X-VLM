```python
i2t_query_ids[:5]

def get_ranks(query_ids, top_ids):
    ranks = []
    for query_id, top_result in zip(query_ids, top_ids):
        if query_id in top_result:
            rank = top_result.index(query_id)
        else:
            rank = 'unk'
        ranks.append(rank)
    return ranks
        
get_ranks()
```
